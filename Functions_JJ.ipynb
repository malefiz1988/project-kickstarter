```python
from IPython.display import display
pd.options.display.max_columns = None
import os
```

```python
df_raw = pd.DataFrame()
for f in os.listdir("./data"):
    if f[:4] == "Kick":
        print(f)
        df = pd.read_csv("./data/" + str(f))
        df_raw = pd.concat([df_raw, df], axis = 0)
```

```python
df_raw = df_raw.drop(columns_drop, axis=1)
df_raw.head()
```

```python
df_suspended = df_raw[df_raw['state']=='suspended']
df_live = df_raw[df_raw['state']=='live']
df_canceled = df_raw[df_raw['state']=='canceled']
df_successful = df_raw[df_raw['state']=='successful']
df_failed = df_raw[df_raw['state']=='failed']
```
